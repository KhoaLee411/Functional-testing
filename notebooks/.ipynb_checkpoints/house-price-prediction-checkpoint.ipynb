```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
DATA_PATH = "../data"
MODEL_PATH = "../models"
```

## EDA

```python
# Read data
train = pd.read_csv(f"{DATA_PATH}/train.csv")
test = pd.read_csv(f"{DATA_PATH}/test.csv")
```

```python
# Drop the column `Id`
train = train.drop(columns=["Id"])
test = test.drop(columns=["Id"])
```

```python
# List of columns to plot
columns_to_plot = set(train.columns) - set("SalePrice")

# Determine number of features to plot
num_features = len(columns_to_plot)
num_rows = (num_features // 5) + int(num_features % 5 != 0)
num_cols = min(5, num_features)  # 5 columns or the number of features if less

# Initialize the figure and axes
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(20, num_rows * 4))
axes = axes.flatten()

# Plot each column, skipping non-numeric ones
for i, column in enumerate(columns_to_plot):
    if column in train.columns:
        # Check if the column is numeric
        if pd.api.types.is_numeric_dtype(train[column]):
            axes[i].scatter(train[column], train["SalePrice"])
            axes[i].set_title(column, fontsize=10)
            axes[i].set_xlabel(column, fontsize=8)
            axes[i].set_ylabel("SalePrice", fontsize=8)
        else:
            # Plot an empty subplot for non-numeric columns
            axes[i].text(
                0.5,
                0.5,
                "Non-Numeric",
                horizontalalignment="center",
                verticalalignment="center",
            )
            axes[i].set_title(column, fontsize=10)
            axes[i].set_xlabel(column, fontsize=8)
            axes[i].set_ylabel("SalePrice", fontsize=8)
    else:
        axes[i].axis("off")  # Turn off the axis if the column is not present

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Remove outliers by rules, you can definitely use other methods such as box plots
train = train.drop(
    train[(train["GrLivArea"] > 4000) & (train["SalePrice"] < 300000)].index
)
train = train.drop(
    train[(train["EnclosedPorch"] > 400) & (train["SalePrice"] > 700000)].index
)
train = train.drop(
    train[(train["1stFlrSF"] > 4000) & (train["SalePrice"] > 400000)].index
)
train = train.drop(
    train[(train["LotArea"] > 100000) & (train["SalePrice"] > 400000)].index
)
```

```python
# Extract the groundtruth column
y_train = train["SalePrice"].values
train = train.drop("SalePrice", axis=1)
```

```python
# Rename some columns starting with a number
replace_dict = {
    "1stFlrSF": "FirstFlrSF",
    "2ndFlrSF": "SecondFlrSF",
    "3SsnPorch": "ThreeSsnPorch",
}
train = train.rename(columns=replace_dict)
test = test.rename(columns=replace_dict)
```

```python
# Measure missing data percentage
missing_count = train.isna().sum()
missing_percentage = (missing_count / len(train)) * 100
missing_summary = pd.DataFrame(
    {"Missing Count": missing_count, "Percentage": missing_percentage}
)
missing_summary_sorted = missing_summary.sort_values(by="Percentage", ascending=False)
print("missing_summary_sorted:")
print(missing_summary_sorted)

# Plot the missing data percentage
plt.figure(figsize=(15, 6))
missing_summary["Percentage"].sort_values().plot(kind="bar", color="skyblue")
plt.title("Percentage of Missing Values per Column")
plt.xlabel("Columns")
plt.ylabel("Percentage of Missing Values")
plt.xticks(ha="right")  # Rotate column names for better readability
plt.show()
```

```python
# Drop columns having missing percentage > 0, remember that it can result in a dramatical decrease
# in the number of rows, so use it with care
columns_to_drop = missing_summary[missing_summary["Percentage"] > 0].index
train = train.drop(columns=columns_to_drop)
test = test.drop(columns=columns_to_drop)
print("Columns dropped:", columns_to_drop.tolist())
print("\nCleaned DataFrame:")
```

```python
# Measure correlations with the target SalePrice
numerical_train = train.select_dtypes(include=["number"])
correlation_matrix = numerical_train.corr()
# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, annot=False, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1
)
plt.title("Correlation Matrix")
plt.show()
```

## Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder

# Convert categorical columns to string type
str_cols = ["MSSubClass", "OverallCond", "YrSold", "MoSold"]
for col in str_cols:
    train[col] = train[col].astype(str)
    test[col] = test[col].astype(str)

# Transform numerical columns that are categorical
num_cat_cols = [
    "ExterQual",
    "ExterCond",
    "HeatingQC",
    "KitchenQual",
    "Functional",
    "LandSlope",
    "LotShape",
    "PavedDrive",
    "Street",
    "CentralAir",
    "MSSubClass",
    "OverallCond",
    "YrSold",
    "MoSold",
    "MSZoning",
    "LandContour",
    "Utilities",
    "LotConfig",
    "Neighborhood",
    "Condition1",
    "Condition2",
    "BldgType",
    "HouseStyle",
    "RoofStyle",
    "RoofMatl",
    "Exterior1st",
    "Exterior2nd",
    "Foundation",
    "Heating",
    "SaleType",
    "SaleCondition",
]

print(test.shape)
print(train.shape)

# Process columns, apply LabelEncoder to categorical features
label_encoders = {}

# Create a combined df from the train and test
combined = pd.concat([train, test], axis=0, ignore_index=True)

for col in num_cat_cols:
    # Define the label encoders
    le = LabelEncoder()
    # Fit to the data
    le.fit(combined[col])
    # And transform
    train[col] = le.transform(list(train[col].values))
    test[col] = le.transform(list(test[col].values))
    # Save the le to a dict
    label_encoders[col] = le
```

## Modeling

```python
import lightgbm as lgb

model = lgb.LGBMRegressor(
    objective="regression",
    num_leaves=5,
    learning_rate=0.05,
    n_estimators=720,
    max_bin=55,
    bagging_fraction=0.8,
    bagging_freq=5,
    feature_fraction=0.2319,
    feature_fraction_seed=9,
    bagging_seed=9,
    min_data_in_leaf=6,
    min_sum_hessian_in_leaf=11,
)

model.fit(train, y_train)
```

## Evaluation

```python
y_pred = model.predict(test)
```

```python
import joblib

# Dump the model and label encoder to joblib
joblib.dump(label_encoders, f"{MODEL_PATH}/label_encoders.joblib")
joblib.dump(model, f"{MODEL_PATH}/model_xgb.joblib")
```
